```python
#zad1
lista = ['gfedcba','gagagaa','abcdefg','abcdefz','abc']
regex = '[a-z]{6}(a|g)'
for i in lista:  
    if(re.search(regex,i)):  
        print(f"{i} zaakceptowano")  
    else:  
        print(f"{i} nie pasuje")  
```

```python
#zad2
kodPocztowy = re.findall("\s[0-9]{2}-[0-9]{3}",text)
print(kodPocztowy)
```

```python
#zad3
phone = re.findall("[0-9]{3}-[0-9]{3}-[0-9]{3}|\S[0-9]{11}",text)
print(phone)
```
